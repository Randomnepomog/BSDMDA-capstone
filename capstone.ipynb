### D.1 Data Extraction Process: 

Upon initial access to the XML file, a significant challenge emerged in incorrect data structuring, which hindered direct parsing. An unplanned step was introduced to the data extraction process to solve this issue. By employing the native Windows program Notepad, I manually restructured the XML file to ensure its appropriate formatting. This involved the implementation of the following adjustments:

* The inclusion of a `<compounds>` tag at the onset of the XML file to encompass all compound-related data.

* Incorporation of an `<?xml>` declaration before the initial `<compound>` tag to specify the XML format and encoding.

* Elimination of redundant `<?xml>` declarations preceding each `<compound>` tag, as they were identified as redundant.

* Addition of a `</compounds>` tag to effectively conclude the encapsulation process.

This strategic intervention was pivotal in enabling the subsequent conversion of the XML file into a DataFrame format, a departure from the initial plan.
To extract data from the adjusted xml file, I utilized the xml.etree.ElementTree library. This technique involved parsing the XML content and extracting the information required for analysis.

```python
tree = ET.parse('toxins.xml')
root = tree.getroot()

# Initialize a list to store data dictionaries
data_list = []
# Iterate over each <compound> element and extract data
for compound_elem in root.findall(".//compound"):
    data_dict = {}
    
    #not-nested tags
    data_dict["version"] = compound_elem.find("version").text
    
    data_dict["creation_date"] = compound_elem.find("creation_date").text
    
    data_dict["update_date"] = compound_elem.find("update_date").text
    
    data_dict["accession"] = compound_elem.find("accession").text
    
    data_dict["common_name"] = compound_elem.find("common_name").text
    
    data_dict["class"] = compound_elem.find("class").text
    
    data_dict["description"] = compound_elem.find("description").text
    
    data_dict["chemical_formula"] = compound_elem.find("chemical_formula").text
    
    data_dict["average_molecular_weight"] = compound_elem.find("average_molecular_weight").text
    
    data_dict["monisotopic_moleculate_weight"] = compound_elem.find("monisotopic_moleculate_weight").text
    
    data_dict["iupac_name"] = compound_elem.find("iupac_name").text
    
    data_dict["traditional_iupac"] = compound_elem.find("traditional_iupac").text
    
    data_dict["cas_registry_number"] = compound_elem.find("cas_registry_number").text
    
    data_dict["smiles"] = compound_elem.find("smiles").text
    
    data_dict["inchi"] = compound_elem.find("inchi").text
    
    data_dict["inchikey"] = compound_elem.find("inchikey").text
    
    data_dict["status"] = compound_elem.find("status").text
    
    data_dict["origin"] = compound_elem.find("origin").text
    
    data_dict["state"] = compound_elem.find("state").text
    
    data_dict["appearance"] = compound_elem.find("appearance").text
    
    data_dict["route_of_exposure"] = compound_elem.find("route_of_exposure").text
    
    data_dict["mechanism_of_toxicity"] = compound_elem.find("mechanism_of_toxicity").text
    
    data_dict["metabolism"] = compound_elem.find("metabolism").text
    
    data_dict["toxicity"] = compound_elem.find("toxicity").text
    
    data_dict["lethaldose"] = compound_elem.find("lethaldose").text
    
    data_dict["carcinogenicity"] = compound_elem.find("carcinogenicity").text
    
    data_dict["use_source"] = compound_elem.find("use_source").text
    
    data_dict["min_risk_level"] = compound_elem.find("min_risk_level").text
    
    data_dict["health_effects"] = compound_elem.find("health_effects").text
    
    data_dict["symptoms"] = compound_elem.find("symptoms").text
    
    data_dict["treatment"] = compound_elem.find("treatment").text
    
    data_dict["drugbank_id"] = compound_elem.find("drugbank_id").text
    
    data_dict["hmdb_id"] = compound_elem.find("hmdb_id").text
    
    data_dict["pubchem_compound_id"] = compound_elem.find("pubchem_compound_id").text
    
    data_dict["chembl_id"] = compound_elem.find("chembl_id").text
    
    data_dict["chemspider_id"] = compound_elem.find("chemspider_id").text
    
    data_dict["kegg_id"] = compound_elem.find("kegg_id").text
    
    data_dict["uniprot_id"] = compound_elem.find("uniprot_id").text
    
    data_dict["omim_id"] = compound_elem.find("omim_id").text
    
    data_dict["chebi_id"] = compound_elem.find("chebi_id").text
    
    data_dict["biocyc_id"] = compound_elem.find("biocyc_id").text
    
    data_dict["ctd_id"] = compound_elem.find("ctd_id").text
    
    data_dict["stitch_id"] = compound_elem.find("stitch_id").text
    
    data_dict["actor_id"] = compound_elem.find("actor_id").text
    
    data_dict["pdb_id"] = compound_elem.find("pdb_id").text
    
    data_dict["wikipidia"] = compound_elem.find("wikipidia").text
    
    
    #nested tags
    # categories should be saved as a list which later can be exploded
    categories_elem = compound_elem.find("categories")
    if categories_elem is not None:
        categories = [category.text for category in categories_elem.findall("category")]
        data_dict["categories"] = categories
    else:
        data_dict["categories"] = []

    # types should be saved as a list which later can be exploded
    types_elem = compound_elem.find("types")
    if types_elem is not None:
        types = [type_one.text for type_one in types_elem.findall("type")]
        data_dict["types"] = types
    else:
        data_dict["types"] = []
    
    # synonyms should be saved as a list which later can be exploded
    synonyms_elem = compound_elem.find("synonyms")
    if synonyms_elem is not None:
        synonyms = [synonym.text for synonym in synonyms_elem.findall("synonym")]
        data_dict["synonyms"] = synonyms
    else:
        data_dict["synonyms"] = []

    # cellular_locations should be saved as a list which later can be exploded
    cellular_locations_elem = compound_elem.find("cellular_locations")
    if cellular_locations_elem is not None:
        cellular_locations = [cellular_location.text for cellular_location in 
                              cellular_locations_elem.findall("cellular_location")]
        data_dict["cellular_locations"] = cellular_locations
    else:
        data_dict["cellular_locations"] = []

    # biofluid_locations should be saved as a list which later can be exploded
    biofluid_locations_elem = compound_elem.find("biofluid_locations")
    if biofluid_locations_elem is not None:
        biofluid_locations = [biofluid_location.text for biofluid_location in 
                              biofluid_locations_elem.findall("biofluid_location")]
        data_dict["biofluid_locations"] = biofluid_locations
    else:
        data_dict["biofluid_locations"] = []

    # tissue_locations should be saved as a list which later can be exploded
    tissue_locations_elem = compound_elem.find("tissue_locations")
    if tissue_locations_elem is not None:
        tissue_locations = [tissue_location.text for tissue_location in tissue_locations_elem.findall("tissue_location")]
        data_dict["tissue_locations"] = tissue_locations
    else:
        data_dict["tissue_locations"] = []
    
    
    # nested tags with nested tags
    # Extract taxonomy information each into its separate field
    taxonomy_elem = compound_elem.find("taxonomy")

    data_dict["kingdom"] = taxonomy_elem.find("kingdom").text
    data_dict["super_class"] = taxonomy_elem.find("super_class").text
    data_dict["taxonomy_class"] = taxonomy_elem.find("class").text
    data_dict["sub_class"] = taxonomy_elem.find("sub_class").text
    data_dict["direct_parent"] = taxonomy_elem.find("direct_parent").text
    data_dict["molecular_framework"] = taxonomy_elem.find("molecular_framework").text

    # nested tags inside taxonomy
    # alternate_parent should be saved as a list which later can be exploded
    alternate_parents_elem = compound_elem.find("alternate_parents")
    if alternate_parents_elem is not None:
        alternate_parents = [alternative_parent.text for alternative_parent in alternate_parents_elem.findall("alternative_parent")]
        data_dict["alternate_parents"] = alternate_parents
    else:
        data_dict["alternate_parents"] = []

    # substituents should be saved as a list which later can be exploded
    substituents_elem = compound_elem.find("substituents")
    if substituents_elem is not None:
        substituents = [substituent.text for substituent in substituents_elem.findall("substituent")]
        data_dict["substituents"] = substituents
    else:
        data_dict["substituents"] = []

    # descriptors should be saved as a list which later can be exploded
    descriptors_elem = compound_elem.find("descriptors")
    if descriptors_elem is not None:
        descriptors = [descriptor.text for descriptor in descriptors_elem.findall("descriptor")]
        data_dict["descriptors"] = descriptors
    else:
        data_dict["descriptors"] = []
    
        
    # Extract pathways information each into its separate field
    pathways_elem = compound_elem.find("pathways")
    if pathways_elem is not None:
        pathways_names = []
        pathways_smpdb_ids = []
        pathways_kegg_map_ids = []
        for pathway_elem in pathways_elem.findall("pathway"):
            pathways_names.append(pathway_elem.find("name").text)
            pathways_smpdb_ids.append(pathway_elem.find("smpdb_id").text)
            pathways_kegg_map_ids.append(pathway_elem.find("kegg_map_id").text)
        data_dict["pathways_names"] = pathways_names
        data_dict["pathways_smpdb_id"] = pathways_smpdb_ids
        data_dict["pathways_kegg_map_id"] = pathways_kegg_map_ids
    else:
        data_dict["pathways_names"] = []
        data_dict["pathways_smpdb_id"] = []
        data_dict["pathways_kegg_map_id"] = []

    # Extract general references information
    general_references_elem = compound_elem.find("general_references")
    if general_references_elem is not None:
        general_references = []
        general_reference_pubmed_ids = []
        for general_reference_elem in general_references_elem.findall("reference"):
            general_references.append(general_reference_elem.find("reference_text").text)
            pubmed_id_elem = general_reference_elem.find("pubmed_id")
            if pubmed_id_elem is not None:
                general_reference_pubmed_ids.append(pubmed_id_elem.text)
            else:
                general_reference_pubmed_ids.append(None)  # or any other suitable value for missing pubmed_id
        data_dict["general_reference_reference_text"] = general_references
        data_dict["general_reference_pubmed_id"] = general_reference_pubmed_ids
    else:
        data_dict["general_reference_reference_text"] = []
        data_dict["general_reference_pubmed_id"] = []
    
    # Extract target information
    targets_elem = compound_elem.find("targets")
    if targets_elem is not None:
        targets_id = []
        targets_name = []
        targets_uniprot_id = []
        targets_mechanism_of_action = []
        targets_references_text = []
        targets_references_pubmed_id = []
        

        for target_elem in targets_elem.findall("target"):
            targets_id.append(target_elem.find("target_id").text)
            targets_name.append(target_elem.find("name").text)
            targets_uniprot_id.append(target_elem.find("uniprot_id").text)
            targets_mechanism_of_action.append(target_elem.find("mechanism_of_action").text)
            

            references = []
            pubmed_ids = []
            target_references_elem = target_elem.find("references")
            if target_references_elem is not None:
                for target_reference_elem in target_references_elem.findall("reference"):
                    references.append(target_reference_elem.find("reference_text").text)
                    pubmed_id_elem = target_reference_elem.find("pubmed_id")
                    if pubmed_id_elem is not None:
                        pubmed_ids.append(pubmed_id_elem.text)
                    else:
                        pubmed_ids.append(None)  # or any other suitable value for missing pubmed_id

            targets_references_text.append(references)
            targets_references_pubmed_id.append(pubmed_ids)
            
            targets_constants_elem = target_elem.find("constants")
            if targets_constants_elem is not None:
                targets_constants = []
                for targets_constant_elem in targets_constants_elem.findall("property"):
                    kind = targets_constant_elem.find("kind").text
                    value = targets_constant_elem.find("value").text
                    targets_constant.append([kind, value])
                data_dict["targets_constants"] = targets_constants
            else:
                data_dict["targets_constants"] = []
                
                

        data_dict["targets_id"] = targets_id
        data_dict["targets_name"] = targets_name
        data_dict["targets_uniprot_id"] = targets_uniprot_id
        data_dict["targets_mechanism_of_action"] = targets_mechanism_of_action
        data_dict["targets_references_text"] = targets_references_text
        data_dict["targets_references_pubmed_id"] = targets_references_pubmed_id
        
    else:
        data_dict["targets_id"] = []
        data_dict["targets_name"] = []
        data_dict["targets_uniprot_id"] = []
        data_dict["targets_mechanism_of_action"] = []
        data_dict["targets_references_text"] = []
        data_dict["targets_references_pubmed_id"] = []
        data_dict["targets_constants"] = []

    # Extract experimental_properties information
    experimental_properties_elem = compound_elem.find("experimental_properties")
    experimental_properties = []
    
    if experimental_properties_elem is not None:
        for experimental_property_elem in experimental_properties_elem.findall("property"):
            kind = experimental_property_elem.find("kind").text
            value = experimental_property_elem.find("value").text
            experimental_properties.append([kind, value])
    
    data_dict["experimental_properties"] = experimental_properties
    
    # Extract predicted properties information
    predicted_properties_elem = compound_elem.find("predicted_properties")
    if predicted_properties_elem is not None:
        predicted_properties = []
        for predicted_property_elem in predicted_properties_elem.findall("property"):
            kind = predicted_property_elem.find("kind").text
            value = predicted_property_elem.find("value").text
            source_elem = predicted_property_elem.find("source")
            source = source_elem.text if source_elem is not None else None
            predicted_properties.append([kind, value, source])
        data_dict["predicted_properties"] = predicted_properties
    else:
        data_dict["predicted_properties"] = []

        
    #Extract spectra information
    spectra_elem = compound_elem.find("spectra")
    spectra = []
    
    if spectra_elem is not None:
        for spectrum_elem in spectra_elem.findall("spectrum"):
            type_spectrum = spectrum_elem.find("type").text
            spectrum_id = spectrum_elem.find("spectrum_id").text
            spectra.append([type_spectrum, spectrum_id])
    
    data_dict["spectra"] = spectra
    
    
    #Save it to the dictionary list
    data_list.append(data_dict)
    
   
```

```python
# Investigate patterns in the amount of missing data in each column.
outlier_columns = []  
def missing_perc_plot(df):
    global outlier_columns
    rows_num = df.shape[0]
    na_values = (df.isna().sum() / rows_num) * 100
    #saving outlier in a list to avoid any mistakes with column names later while dropping
    q3= na_values.quantile(0.75)
    q1 = na_values.quantile(0.25)
    threshold = q3+1.5*(q3-q1)
    for column, percentage in na_values.items():
        if percentage > threshold:
            outlier_columns.append(column)
    
    sns.set_style("darkgrid", {"axes.facecolor": "lightgrey"})
    plt.figure(figsize=(10, 20))
    ax = sns.barplot(x=na_values.values, 
                     y=na_values.index,
                     palette=sns.color_palette("viridis_r", 
                     n_colors=len(na_values), as_cmap=True)(na_values.values))

    plt.title("Percentage of Missing Values in Each Column")
    plt.xlabel("Missing Values, %")
    plt.ylabel("Columns")
    plt.xticks(rotation=0)
    plt.grid(True)

    # Add numbers near each bar
    for i, v in enumerate(na_values.values):
        ax.text(v + 1, i, f"{v:.2f}", color='black', va='center')
   

    plt.show()
    print(threshold)
    


missing_perc_plot(df)
f"Outlier columns:{outlier_columns}"
    
```

```python
# function to restore missing values
def restore_missing_vals(row, col1, col2):
    if pd.isna(row[col1]) and not pd.isna(row[col2]):
        return row[col2]
    else:
        return row[col1]
```

```python
from bs4 import BeautifulSoup

# path
html_file_path = 'boilmelt.html'

# read the HTML content from the file
with open(html_file_path, 'r', encoding='utf-8') as file:
    html = file.read()
    
    
# create a BeautifulSoup object from the HTML
soup = BeautifulSoup(html, "html.parser")

# find all table rows within the tbody elements
table_rows = soup.find_all("tr")

# create an empty list to store the data
data = []

# iterate through each table row
for row in table_rows:
    # extract data from the table cells within each row
    cells = row.find_all("td")

    if len(cells) >= 4:
        name = cells[0].text.strip()
        formula = cells[1].text.strip()
        boiling_point = cells[2].text.strip()
        melting_point = cells[3].text.strip()

        if boiling_point == "":
            boiling_point = None
        if melting_point == "":
            melting_point = None

        data.append({"common_name": name, "chemical_formula": formula, "boiling_point": boiling_point, "melting_point": melting_point})

# create a DataFrame from the extracted data
df_bm_points = pd.DataFrame(data)


df_bm_points
```

```python
def train_predict(learner, X_train, y_train, X_test, y_test):
    results={}


    # Fit the learner to the training data
    start = time() # Get start time
    learner = learner.fit(X_train, y_train)
    end = time() # Get end time

    # Calculate the training time
    results['train_time'] = end - start

    # Get the predictions on the test set (X_test)
    start = time() # Get start time
    predictions_test = learner.predict(X_test)
    predictions_train = learner.predict(X_train)
    end = time() # Get end time

    # Calculate the total prediction time
    results['pred_time'] = end - start

    # Compute accuracy on training and test sets
    results['acc_train'] = accuracy_score(y_train, predictions_train)
    results['acc_test'] = accuracy_score(y_test, predictions_test)

    # Compute F-score on training and test sets
    results['f_train'] = fbeta_score(y_train, predictions_train, beta=0.5, average='micro')
    results['f_test'] = fbeta_score(y_test, predictions_test, beta=0.5, average='micro')

    # Return the results
    return results
```

```python
#naive prediction
# Counting the ones for each label
TP = np.sum(y, axis=0)
FP = y.shape[0] - TP
TN = 0  # No predicted negatives in the naive case
FN = 0  # No predicted negatives in the naive case

# Calculate accuracy, precision, and recall for each label
accuracy = (TP + TN) / (TP + FP + TN + FN)
recall = TP / (TP + FN)
precision = TP / (TP + FP)

# Calculate F-score for each label using the formula above for beta = 0.5
beta = 0.5
fscore = (1 + beta**2) * (precision * recall) / ((beta**2 * precision) + recall)

# Calculate average accuracy, precision, and recall over all labels
accuracy = np.mean(accuracy)
fscore = np.mean(fscore)


# Print the results
print("Naive Predictor: [Accuracy score: {:.4f}, F-score: {:.4f}]".format(accuracy, fscore))
```
